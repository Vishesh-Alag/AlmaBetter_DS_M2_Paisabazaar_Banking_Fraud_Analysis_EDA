````python




# Standardize missing values (Convert "No Data" & "Not Specified" to NaN)
data["Type_of_Loan"] = data["Type_of_Loan"].replace(["No Data", "Not Specified"], pd.NA)

# Clean delimiters
data["Type_of_Loan"] = data["Type_of_Loan"].str.replace(" and ", ",", regex=False).str.replace(", ", ",", regex=False)

# Function to clean loan types (Remove "Not Specified", duplicates, extra spaces, keep NaN)
def clean_loan_types(loan_str):
    if pd.isna(loan_str):  # If value is NaN, return as is
        return loan_str
    loan_list = [loan.strip() for loan in loan_str.split(",") if loan.strip() and loan.strip() != "Not Specified"]
    return ",".join(sorted(set(loan_list))) if loan_list else pd.NA  # Remove duplicates, keep NaN if empty

data["Type_of_Loan"] = data["Type_of_Loan"].apply(clean_loan_types)

data.to_csv("processed_paisabazaar_dataset.csv", index=False)
````

## **Let's Perform Some Statistical Test like Hypothesis Testing or ANNOVA Test**

````python
df= data2
````

###**ANNOVA TEST 1 : Is there a significant difference in Num_of_Delayed_Payment across Credit_Score categories?**

**This is super relevant because:**

  * More delays = potential risk

  * Lower credit score customers are expected to delay more

````python
from scipy.stats import f_oneway

# Drop missing values in the relevant columns
# Grouping delayed payments by Credit_Score
groups = df.groupby('Credit_Score')['Num_of_Delayed_Payment'].apply(list)

# Extract lists
delay_lists = list(groups)

# Run ANOVA
f_stat, p_value = f_oneway(*delay_lists)

print("F-statistic:", f_stat)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis â€” Significant difference in delays across Credit Score groups.")
else:
    print("Fail to reject the null hypothesis â€” No significant difference.")
````

````python
import pandas as pd
import scipy.stats as stats

# Load the dataset

# Assuming 'Credit_Score' has categories like 'Poor', 'Standard', 'Good'
# Separate the monthly balance data for each credit score category
monthly_balance_poor = data2[data2['Credit_Score'] == 'Poor']['Monthly_Balance']
monthly_balance_standard = data2[data2['Credit_Score'] == 'Standard']['Monthly_Balance']
monthly_balance_good = data2[data2['Credit_Score'] == 'Good']['Monthly_Balance']

# Perform ANOVA test
f_statistic, p_value = stats.f_oneway(monthly_balance_poor, monthly_balance_standard, monthly_balance_good)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the mean monthly balance among credit score categories.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean monthly balance among credit score categories.")
````
